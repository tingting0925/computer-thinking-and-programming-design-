```
x = float(input('請輸入該點的X座標?'))
y = float(input('請輸入該點的Y座標?'))

if (x > 0):
  if (y > 0):
    print('該點在第一象限')
  else:
    print('該點在第四象限')
else:
  if (y > 0):
    print('該點在第二象限')
  else:
    print('該點在第三象限')
```

```
a = int(input('請輸入三角形邊長a長度為?'))
b = int(input('請輸入三角形邊長b長度為?'))
c = int(input('請輸入三角形邊長c長度為?'))

if (a<b+c)and(b<a+c)and(c<a+b):
    print('可構成三角形')
else:
    print('無法構成三角形')
```

```
w = float(input('請輸入體重(KG)?'))
h = float(input('請輸入身高(M)'))
bmi = w/(h*h)
print('BMI為', bmi)

if bmi < 18:
    print('體重過輕')
elif bmi < 24:
    print('體重正常')
elif bmi < 27:
    print('體重過重')
else:
    print('體重肥胖ˇ')
```
