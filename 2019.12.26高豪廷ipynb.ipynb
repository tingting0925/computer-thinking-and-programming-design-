```
#請寫一個程式模擬密碼登入，使用者輸入密碼，
#若密碼一致，則輸出「密碼正確」，否則輸出「登入失敗」 TRY AGAIN
answer = "123"
while True:
    password = input("Enter Your Password: ")
    if password == answer:
        print("Correct")
        break
    else:
        print("Fail, Try Again")
```

```
#請寫一個程式模擬密碼登入，使用者輸入密碼，
#若密碼一致，則輸出「密碼正確」，否則輸出「登入失敗」
answer = "0925"
while True:
    account = input("Enter Your Account: ")
    password = input("Enter Your Password: ")
    if account == answer:
        print("Correct")
        break
    else:
        print("Fail, Try Again")
    if password == answer:
        print("Correct")
        break
    else:
        print("Fail, Try Again")
```

```
password = "09252509"
times = 0

while times <= 3:
    answer = input("Enter Your Password: ")
    if answer == password:
        print("Correct")
        break
    else:
        times = times + 1
        chance = 3 - times
        print("Fail, Try Again", chance, "chance")
```
